## 1. Python solution1

```python
board = [[0]*101 for _ in range(101)]

# 0: (), 1: (), 2: (), 3: ()
dx = [1, 0, -1, 0]
dy = [0, -1, 0, 1]

for _ in range(int(input())):
    x, y, d, g = map(int, input().split(' '))
    board[x][y] = 1
    temp = [d]
    move = [d] #  
    for _ in range(g + 1):
        for k in move:
            x += dx[k]
            y += dy[k]
            board[x][y] = 1
        move = [(i + 1) % 4 for i in temp]
        move.reverse()
        temp = temp + move
    
answer = 0
for i in range(100):
    for j in range(100):
        if (board[i][j] and board[i+1][j] and board[i][j+1] and board[i+1][j+1]):
            answer += 1
print(answer)
```
