## 1. Python solution1

```python
import heapq

N = int(input()) #  
array = []
q = []

for _ in range(N):
    x, y = map(int, input().split())
    array.append((x, y))
array.sort()

for i in array:
    a = i[0] # 
    heapq.heappush(q, i[1])
    #      
    if a < len(q):
        heapq.heappop(q)
    
print(sum(q))
```

```python
import heapq

N = int(input()) #  
array = [list(map(int, input().split())) for _ in range(N)]
array.sort()

result = []
for deadline, count in array:
    if deadline == len(result):
        if count <= result[0]:
            continue
        heapq.heappop(result)
    heapq.heappush(result, count)
    
print(sum(result))
```
